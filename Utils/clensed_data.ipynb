```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
import contractions
from googletrans import Translator

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
def clean_data(data):
  if isinstance(data,str):
    data = contractions.fix(data)
    data = re.sub(r'[^a-zA-Z\s]','',data)
    data = data.lower().strip()
    tokens = word_tokenize(data)
    stop_words = set(stopwords.words('english'))
    stop_words.discard('not')
    tokens = [word for word in tokens if word not in stop_words]

    cleaned_data = ' '.join(tokens)
    return cleaned_data
```

```python
data = pd.read_csv("/content/4_7_stars_kluska_krakow.csv")
df = pd.DataFrame(data)
df.head()
```

```python
translator = Translator()

df['translate'] = df['wiI7pd'].apply(lambda x: translator.translate(x, src='auto', dest='en').text)
```

```python
df.head()
```

```python
df['translate'] = df['translate'].str.replace(r'[.!]', '. ', regex=True)
```

```python
df['text'] = df['translate'].apply(clean_data)
```

```python
df = df.drop(df[(df['translate'] == 'Into') | (df['translate'] == ' ')].index)
df.shape[0]
```

```python
df.to_csv('4_7_review.csv',index=False)
```
